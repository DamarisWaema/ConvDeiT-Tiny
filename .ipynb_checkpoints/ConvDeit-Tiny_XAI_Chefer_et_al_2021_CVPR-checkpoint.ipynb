# Code to classify uploaded image and provide explanation using gradient-weighted transformer attribution as proposed in Chefer et al (2021) - CVPR

```python
import torch
import torch.nn as nn
from torchvision import transforms
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from timm.models import create_model
from timm.models.vision_transformer import Block
import torch.nn.functional as F
from timm.models.layers import DropPath
import cv2
import os
# Seed setting
import random
import json
# -----------------------------
# 1 Device
# -----------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# === Reproducibility ===
def set_seed(seed: int = 42, save_path: str = "seed_log.json"):
    """
    Set the random seed for Python, NumPy, and PyTorch.
    Saves the seed to a file so it can be reused later.
    """
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

    # Save seed info
    seed_data = {
        "seed": seed
    }

    with open(save_path, "w") as f:
        json.dump(seed_data, f)

    #print(f"[INFO] Seed set to {seed} and logged in '{save_path}'.")
set_seed (42)
# -----------------------------
# 2 Custom DeiT-Tiny Definition (same as before)
# -----------------------------
class DepthwiseConv(nn.Module):
    def __init__(self, dim, kernel_size=3):
        super().__init__()
        self.dwconv = nn.Conv2d(dim, dim, kernel_size=kernel_size,
                                padding=kernel_size // 2, groups=dim)
        self.bn = nn.BatchNorm2d(dim)

    def forward(self, x):
        B, N, C = x.shape
        cls_token = x[:, :1]
        patch_tokens = x[:, 1:]
        H = W = int(patch_tokens.shape[1] ** 0.5)
        patch_tokens = patch_tokens.transpose(1, 2).view(B, C, H, W)
        patch_tokens = self.bn(self.dwconv(patch_tokens))
        patch_tokens = patch_tokens.flatten(2).transpose(1, 2)
        return torch.cat((cls_token, patch_tokens), dim=1)

class ParallelBlock(Block):
    def __init__(self, dim, num_heads, mlp_ratio=4., qkv_bias=True,
                 norm_layer=nn.LayerNorm, drop_path=0.0):
        super().__init__(dim, num_heads, mlp_ratio, qkv_bias=qkv_bias,
                         norm_layer=norm_layer, drop_path=0.0)
        self.depthwise_conv = DepthwiseConv(dim)
        self.attn_norm = norm_layer(dim)
        self.conv_norm = norm_layer(dim)
        self.fuse_mlp = nn.Sequential(
            nn.Linear(dim * 2, dim * 4),
            nn.GELU(),
            nn.Linear(dim * 4, dim)
        )
        self.drop_path = DropPath(drop_path) if drop_path > 0. else nn.Identity()

    def forward(self, x):
        x_attn = self.attn(self.attn_norm(x))
        x_conv = self.depthwise_conv(self.conv_norm(x))
        x_fused = torch.cat([x_attn, x_conv], dim=-1)
        x_fused = self.fuse_mlp(x_fused)
        x = x + self.drop_path(x_fused)
        x = x + self.drop_path(self.mlp(self.norm2(x)))
        return x

def load_custom_parallel_deit(num_classes, pretrained=False, replace_blocks=3):
    model = create_model('deit_tiny_patch16_224', pretrained=pretrained)
    for i, block in enumerate(model.blocks):
        if i < replace_blocks:
            mlp_hidden = block.mlp.fc1.out_features
            mlp_in = block.mlp.fc1.in_features
            mlp_ratio = mlp_hidden / mlp_in
            model.blocks[i] = ParallelBlock(
                dim=block.norm1.normalized_shape[0],
                num_heads=block.attn.num_heads,
                mlp_ratio=mlp_ratio,
                qkv_bias=block.attn.qkv.bias is not None,
                norm_layer=type(block.norm1),
                drop_path=0.0
            )
    return model

# -----------------------------
# 3 Load model
# -----------------------------
# Replace with your actual class names

#class_names = ['Fall armyworm', 'Healthy', 'Northern Leaf Blight']  
class_names = ['Blight', 'Grey Leaf Spot', 'Healthy', 'Common Rust']


model = load_custom_parallel_deit(num_classes=len(class_names), pretrained=False, replace_blocks=3)
model.head = nn.Linear(model.head.in_features, len(class_names))
#model.load_state_dict(torch.load("Primary_pretrained_gridsearch_convDeiTTiny.pth", map_location=device))

model.load_state_dict(torch.load("COMLDD_pretrained_gridsearch_convDeiTTiny.pth", map_location=device))

model.to(device).eval()
#print(model)
# -----------------------------
# 4 Transform 
# -----------------------------
transform_test = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])


def apply_colormap_on_image(org_img, activation, alpha=0.45):
    """
    org_img: numpy RGB float32 (0..1)
    activation: numpy float32 heatmap 0..1
    """
    act_uint8 = (activation * 255).astype(np.uint8)
    heatmap = cv2.applyColorMap(act_uint8, cv2.COLORMAP_JET)[:, :, ::-1] / 255.0
    overlay = (1 - alpha) * org_img + alpha * heatmap
    return np.clip(overlay, 0, 1)

#XAI
import types

def enable_chefer_attribution(model):

    for blk in model.blocks:

        attn_module = blk.attn

        def forward_with_hooks(self, x):

            B, N, C = x.shape

            qkv = self.qkv(x)
            qkv = qkv.reshape(B, N, 3, self.num_heads, C // self.num_heads)
            qkv = qkv.permute(2, 0, 3, 1, 4)

            q, k, v = qkv[0], qkv[1], qkv[2]

            attn = (q @ k.transpose(-2, -1)) * self.scale
            attn = attn.softmax(dim=-1)

            # -------------------------------------------------
            #  STORE ATTENTION
            # -------------------------------------------------
            self.attn_map = attn

            # -------------------------------------------------
            #  STORE GRADIENTS OF ATTENTION
            # -------------------------------------------------
            attn.register_hook(lambda grad: setattr(self, "attn_grad", grad))

            attn = self.attn_drop(attn)

            x = (attn @ v).transpose(1, 2).reshape(B, N, C)
            x = self.proj(x)
            x = self.proj_drop(x)

            return x

        attn_module.forward = types.MethodType(
            forward_with_hooks,
            attn_module
        )
        
def chefer_attribution(model, input_tensor, target_class=None):

    model.zero_grad()

    logits = model(input_tensor)

    if target_class is None:
        target_class = logits.argmax(dim=-1)

    loss = logits[0, target_class]
    loss.backward()

    # -------------------------------------------------
    # Collect maps + grads
    # -------------------------------------------------
    attn_maps = [blk.attn.attn_map for blk in model.blocks]
    attn_grads = [blk.attn.attn_grad for blk in model.blocks]

    # Use first image in batch
    attn_maps = [a[0] for a in attn_maps]
    attn_grads = [g[0] for g in attn_grads]

    num_tokens = attn_maps[0].size(-1)

    R = torch.eye(num_tokens).to(input_tensor.device)

    # -------------------------------------------------
    # Relevance propagation
    # -------------------------------------------------
    for attn, grad in zip(reversed(attn_maps), reversed(attn_grads)):

        # Average heads
        attn = attn.mean(dim=0)
        grad = grad.mean(dim=0)

        # Gradient-weighted attention
        cam = (attn * grad).clamp(min=0)

        # Add residual
        cam = cam + torch.eye(num_tokens).to(cam.device)

        cam = cam / cam.sum(dim=-1, keepdim=True)

        R = cam @ R

    # -------------------------------------------------
    # CLS  patches
    # -------------------------------------------------
    patch_rel = R[0, 1:]

    patch_rel = patch_rel - patch_rel.min()
    patch_rel = patch_rel / (patch_rel.max() + 1e-8)

    grid_size = int(patch_rel.shape[0] ** 0.5)

    heatmap = patch_rel.view(grid_size, grid_size)

    heatmap = F.interpolate(
        heatmap.unsqueeze(0).unsqueeze(0),
        size=(224, 224),
        mode='bilinear',
        align_corners=False
    ).squeeze().detach().cpu().numpy()

    return heatmap, logits
# ============================================================
# MAIN EXPLANATION FUNCTION 
# ============================================================

def explain_image(image_path, save_dir="XAI_Explanations"):
    
    device = next(model.parameters()).device
    os.makedirs(save_dir, exist_ok=True)

    # Preprocessing
    transform_test = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.5]*3, [0.5]*3)
    ])

    img = Image.open(image_path).convert("RGB")
    img_224 = img.resize((224, 224))
    rgb_np = np.array(img_224).astype(np.float32) / 255.0

    tensor_img = transform_test(img).unsqueeze(0).to(device)


    model.eval()
    enable_chefer_attribution(model)

    heatmap, logits = chefer_attribution(model, tensor_img)
    
    
    # -----------------------------
    # Prediction
    # -----------------------------
    probs = torch.softmax(logits, dim=1)[0]
    pred_idx = int(torch.argmax(probs))
    pred_name = class_names[pred_idx]
    confidence = float(probs[pred_idx])

    # -----------------------------
    # Overlay heatmap
    # -----------------------------
    overlay = apply_colormap_on_image(rgb_np, heatmap, alpha=0.5)
    
    
    
    # Save
    save_path = os.path.join(save_dir, f"XAI_{os.path.basename(image_path)}")
    cv2.imwrite(save_path, (overlay * 255).astype(np.uint8)[:, :, ::-1])

    # Display
    plt.figure(figsize=(10, 5))
    plt.suptitle(f"{pred_name}  (Confidence: {confidence:.2f})", fontsize=14)

    plt.subplot(1, 2, 1)
    plt.imshow(img_224)
    plt.title("Original", fontsize=10)
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(overlay)
    plt.title(f"{pred_name}  (Confidence: {confidence*100:.2f}%)", fontsize=10)
    plt.axis("off")

    plt.show()

    return pred_name, confidence, save_path

#pred_name, conf, path = explain_image("/notebooks/Primary_Dataset/test/faw/corn_faw_1037.jpg")
pred_name, conf, path = explain_image("/notebooks/COMLDD_Dataset/test/blight/Corn_Blight (311).JPG")

print("Prediction:", pred_name)
print("Confidence:", conf)

print("Saved at:", path)
```
