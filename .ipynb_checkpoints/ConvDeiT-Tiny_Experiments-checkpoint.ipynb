```python
import timm
import torch
from torch import nn, optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from tqdm import tqdm
import os
import random
import numpy as np
from PIL import Image, ImageFilter
from torchvision.transforms import ColorJitter
from torch.optim.lr_scheduler import CosineAnnealingLR
# Customizing DEIT-Ti
import torch
import torch.nn as nn
from timm.models.vision_transformer import VisionTransformer, Block
from timm.models import create_model
from functools import partial
from PIL import Image
#for CIFAR-Dataset
from torchvision.datasets import CIFAR100
from torch.utils.data import random_split, ConcatDataset
import torchvision.transforms.functional as F
from datasets import load_dataset # for imagenet100 dataset
# Seed setting
import random
import numpy as np
import torch
import os
import json

# === Reproducibility ===
def set_seed(seed: int = 42, save_path: str = "seed_log.json"):
    """
    Set the random seed for Python, NumPy, and PyTorch.
    Saves the seed to a file so it can be reused later.
    """
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

    # Save seed info
    seed_data = {
        "seed": seed
    }

    with open(save_path, "w") as f:
        json.dump(seed_data, f)

    print(f"[INFO] Seed set to {seed} and logged in '{save_path}'.")
set_seed (42)

# === Set device ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

#MODEL PARAMETER COUNTING
def count_params(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

#Code for noise addition
class AddGaussianNoise:
    def __init__(self, mean=0.0, std=0.05):
        self.mean = mean
        self.std = std

    def __call__(self, img):
        if isinstance(img, Image.Image):
            img = np.array(img).astype(np.float32) / 255.0
        noise = np.random.normal(self.mean, self.std, img.shape)
        noisy_img = np.clip(img + noise, 0, 1)
        return Image.fromarray((noisy_img * 255).astype(np.uint8))

class RandomNoiseBlurOrJitter:
    def __init__(self, prob=0.2, noise_std=0.05, blur_radius=1, jitter_params=None):
        self.prob = prob
        self.noise_transform = AddGaussianNoise(std=noise_std)
        self.blur_radius = blur_radius
        self.jitter_transform = ColorJitter(**(jitter_params or {
            'brightness': 0.2,
            'contrast': 0.2,
            'saturation': 0.2,
            'hue': 0.1
        }))

    def __call__(self, img):
        if random.random() < self.prob:
            choice = random.choice(['noise', 'blur', 'jitter'])
            if choice == 'noise':
                img = self.noise_transform(img)
            elif choice == 'blur':
                img = img.filter(ImageFilter.GaussianBlur(radius=self.blur_radius))
            elif choice == 'jitter':
                img = self.jitter_transform(img)
        return img


# === Data transforms ===
# === Define CIFAR-100 Transforms ===
base_transform_train = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomRotation(270),
    transforms.RandomHorizontalFlip(p=0.5),
    RandomNoiseBlurOrJitter(prob=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

base_transform_val = transforms.Compose([
    transforms.Resize((224, 224)),
    RandomNoiseBlurOrJitter(prob=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

base_transform_test = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

def add_horizontal_flips(dataset):
    flipped_data = []
    for img, label in dataset:
        flipped_img = F.hflip(img)
        flipped_data.append((flipped_img, label))
    return ConcatDataset([dataset, flipped_data])
'''
# === Load CIFAR-100 Training Set ===
full_train_dataset = CIFAR100(root='./data', train=True, download=True, transform=base_transform_train)

# === Split into 90% train, 10% val ===
train_size = int(0.9 * len(full_train_dataset))
val_size = len(full_train_dataset) - train_size
train_dataset, val_dataset = random_split(full_train_dataset, [train_size, val_size])


# === Load Test Set ===
test_dataset = CIFAR100(root='./data', train=False, download=True, transform=base_transform_test)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=2)

'''
'''
# === Load ImageNet-100 from Hugging Face (Streaming) ===


from datasets import load_dataset
from torch.utils.data import IterableDataset, DataLoader
import torch

# === Load ImageNet-100 (streaming) ===
hf_train = load_dataset("clane9/imagenet-100", split="train", streaming=True)
hf_val   = load_dataset("clane9/imagenet-100", split="validation", streaming=True)

# --- Custom IterableDataset wrapper ---
class HFImageNetIterable(IterableDataset):
    def __init__(self, hf_dataset, transform):
        self.hf_dataset = hf_dataset
        self.transform = transform

    def __iter__(self):
        for example in self.hf_dataset:
            image = self.transform(example["image"].convert("RGB"))
            label = torch.tensor(example["label"], dtype=torch.long)
            yield image, label

# === Wrap datasets with transforms ===
train_dataset = HFImageNetIterable(hf_train.shuffle(buffer_size=10_000), base_transform_train)
val_dataset   = HFImageNetIterable(hf_val, base_transform_val)

# === DataLoaders (num_workers=0 for streaming) ===
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=False, num_workers=0)
val_loader   = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=0)

print("‚úÖ ImageNet-100 streaming loaders ready")

#End of imagenet 100 dataset code
'''
# === Load datasets ===

# tiny imagenet 200 

import os
import shutil

# New local working folder - make sure this is writable!
local_val_dir = "/notebooks/tiny-imagenet-200/val"
os.makedirs(local_val_dir, exist_ok=True)

# Copy val_annotations.txt
shutil.copy("/datasets/tiny-imagenet-200/val/val_annotations.txt", local_val_dir)

# Copy val images folder recursively
shutil.copytree("/datasets/tiny-imagenet-200/val/images", os.path.join(local_val_dir, "images"))

# Now, run the organizing code on local_val_dir
annotations_file = os.path.join(local_val_dir, "val_annotations.txt")
val_images_dir = os.path.join(local_val_dir, "images")

with open(annotations_file, "r") as f:
    data = f.readlines()

img_to_class = {}
for line in data:
    parts = line.strip().split('\t')
    img_filename, class_label = parts[0], parts[1]
    img_to_class[img_filename] = class_label

for class_label in set(img_to_class.values()):
    class_dir = os.path.join(local_val_dir, class_label)
    os.makedirs(class_dir, exist_ok=True)

for img_filename, class_label in img_to_class.items():
    src_path = os.path.join(val_images_dir, img_filename)
    dst_path = os.path.join(local_val_dir, class_label, img_filename)
    if os.path.exists(src_path):
        shutil.move(src_path, dst_path)
    else:
        print(f"Warning: {src_path} not found")

# Clean up the empty images directory
if os.path.exists(val_images_dir) and not os.listdir(val_images_dir):
    os.rmdir(val_images_dir)

# === Paths (mounted dataset) ===
train_dir = "/datasets/tiny-imagenet-200/train"


# === Load datasets ===
train_dataset = datasets.ImageFolder(train_dir, transform=base_transform_train)
val_dataset   = datasets.ImageFolder(local_val_dir, transform=base_transform_val)

# === Dataloaders ===
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
val_loader   = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)

# Check
print(f"Train samples: {len(train_dataset)}")
print(f"Val samples:   {len(val_dataset)}")
print(f"Classes:       {len(train_dataset.classes)}")
print(train_loader.dataset.class_to_idx)
print(val_loader.dataset.class_to_idx)

# end of tiny imagenet 200

# Adding depthwise convolutions after only the first 2 MHSA blocks
import torch
import torch.nn as nn
from timm.models import create_model
from timm.models.vision_transformer import Block
from timm.models.vision_transformer import DropPath

# 1. Depthwise convolution module with normalization
class DepthwiseConv(nn.Module):
    def __init__(self, dim, kernel_size=3):
        super().__init__()
        self.dwconv = nn.Conv2d(dim, dim, kernel_size=kernel_size, padding=kernel_size // 2, groups=dim)
        self.bn = nn.BatchNorm2d(dim)

    def forward(self, x):
        B, N, C = x.shape
        cls_token = x[:, :1]
        patch_tokens = x[:, 1:]

        H = W = int(patch_tokens.shape[1] ** 0.5)
        patch_tokens = patch_tokens.transpose(1, 2).view(B, C, H, W)
        patch_tokens = self.bn(self.dwconv(patch_tokens))
        patch_tokens = patch_tokens.flatten(2).transpose(1, 2)

        return torch.cat((cls_token, patch_tokens), dim=1)

# 2. Custom block with parallel MHSA and DepthwiseConv

# concat+,mlp+ residual -- Uses residual connections around the fused output and the MLPlatest -- Best

class ParallelBlock(Block):
    def __init__(self, dim, num_heads, mlp_ratio=4., qkv_bias=True, norm_layer=nn.LayerNorm, drop_path=0.0):
        super().__init__(dim, num_heads, mlp_ratio, qkv_bias=qkv_bias, norm_layer=norm_layer, drop_path=0.0)

        self.depthwise_conv = DepthwiseConv(dim)
        self.attn_norm = norm_layer(dim)
        self.conv_norm = norm_layer(dim)

        # üîÑ Fusion MLP
        self.fuse_mlp = nn.Sequential(
            nn.Linear(dim * 2, dim * 4),
            nn.GELU(),
            nn.Linear(dim * 4, dim)
        )

        

        self.drop_path = DropPath(drop_path) if drop_path > 0. else nn.Identity()

    def forward(self, x):
        # Parallel MHSA and depthwise conv paths
        x_attn = self.attn(self.attn_norm(x))
        x_conv = self.depthwise_conv(self.conv_norm(x))

        # Fusion
        x_fused = torch.cat([x_attn, x_conv], dim=-1)  # (B, N, 2C)
        x_fused = self.fuse_mlp(x_fused)               # (B, N, C)

      

        # Residual + MLP
        x = x + self.drop_path(x_fused)
        x = x + self.drop_path(self.mlp(self.norm2(x)))

        return x

# 3. Load model and selectively replace first 3 blocks
def load_custom_deit(num_classes=200, pretrained=True, freeze_pretrained=True, replace_blocks=3):
    model = create_model('deit_tiny_patch16_224', pretrained=pretrained)

    modified_blocks = []

    for i, block in enumerate(model.blocks):
        if i < replace_blocks:
            mlp_hidden = block.mlp.fc1.out_features
            mlp_in = block.mlp.fc1.in_features
            mlp_ratio = mlp_hidden / mlp_in

            new_block = ParallelBlock(
                dim=block.norm1.normalized_shape[0],
                num_heads=block.attn.num_heads,
                mlp_ratio=mlp_ratio,
                qkv_bias=block.attn.qkv.bias is not None,
                norm_layer=type(block.norm1),
                drop_path=0.0
            )

            # Copy over pretrained weights
            new_block.norm1.load_state_dict(block.norm1.state_dict())
            new_block.attn.load_state_dict(block.attn.state_dict())
            new_block.norm2.load_state_dict(block.norm2.state_dict())
            new_block.mlp.load_state_dict(block.mlp.state_dict())

            # Freeze only the inherited parts
            for name, param in new_block.named_parameters():
                if name.startswith("attn") or name.startswith("norm1") or name.startswith("norm2")  or name.startswith("mlp"):
                    param.requires_grad = False
                else:
                    param.requires_grad = True  # fuse_mlp, depthwise_conv, etc.

            model.blocks[i] = new_block
            modified_blocks.append(new_block)

    # Replace head
    model.head = nn.Linear(model.head.in_features, num_classes)

    if freeze_pretrained:
        # First freeze the whole model
        for param in model.parameters():
            param.requires_grad = False

        # Re-enable head and modified block components
        for param in model.head.parameters():
            param.requires_grad = True

        for block in modified_blocks:
            for name, param in block.named_parameters():
                if not (name.startswith("attn") or name.startswith("norm1") or name.startswith("norm2") or name.startswith("mlp")):
                    param.requires_grad = True


    def print_trainable_parameters(model):
        print("Trainable parameters:")
        for name, param in model.named_parameters():
            if param.requires_grad:
                print(f"‚úÖ {name} ‚Äî {param.numel():,} parameters")

    print_trainable_parameters (model)
    
    return model

   

# Instantiate and move to device
model = load_custom_deit(num_classes=200, pretrained=True, freeze_pretrained=True, replace_blocks=3)
print("\n‚úÖ Trainable parameters when freezing pretrained parts:")
for name, param in model.named_parameters():
    if param.requires_grad:
        print(name)
model.to(device)
print(f"Total Trainable Parameters in ConvDeiT-Tiny model is: {count_params(model):,}")


# === Loss and optimizer ===
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=0)
scheduler = None


# === Training with Early Stopping ===
best_acc = 0
epochs = 200
patience = 20
patience_counter = 0
best_epoch = 0

for epoch in range(epochs):
    model.train()
    running_loss = 0

    for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=False):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    if scheduler:
        scheduler.step()

    # === Validation ===
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    acc = correct / total
    avg_loss = running_loss / len(train_loader)
    print(f"Epoch {epoch+1:>3} ‚Äî Loss: {avg_loss:.4f}, Val Accuracy: {acc*100:.2f}%")

    # === Check for improvement ===
    if acc > best_acc:
        best_acc = acc
        best_epoch = epoch + 1
        patience_counter = 0
        torch.save(model.state_dict(), 'TinyImageNet200_pretrained_ConvDeiT-Tiny.pth')

        print("‚úÖ New best model saved.")
    else:
        patience_counter += 1
        print(f"‚è≥ Patience Counter: {patience_counter}/{patience}")
        if patience_counter >= patience:
            print("üõë Early stopping triggered.")
            break

print(f"\nüéØ Best validation accuracy  : {best_acc*100:.2f}% at Epoch {best_epoch}")

```

```python
import timm
import torch
from torch import nn, optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from tqdm import tqdm
import os
import random
import numpy as np
from PIL import Image, ImageFilter
from torchvision.transforms import ColorJitter
from torch.optim.lr_scheduler import CosineAnnealingLR
from itertools import product
import time
# Customizing DEIT Tiny
import torch
import torch.nn as nn
from timm.models.vision_transformer import VisionTransformer, Block
from timm.models import create_model
from functools import partial
import torch.nn.functional as F
#Flops and efficiency logging
import time
from ptflops import get_model_complexity_info


# Seed setting
import random
import numpy as np
import torch
import os
import json

# === Reproducibility ===
def set_seed(seed: int = 42, save_path: str = "seed_log.json"):
    """
    Set the random seed for Python, NumPy, and PyTorch.
    Saves the seed to a file so it can be reused later.
    """
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

    # Save seed info
    seed_data = {
        "seed": seed
    }

    with open(save_path, "w") as f:
        json.dump(seed_data, f)

    print(f"[INFO] Seed set to {seed} and logged in '{save_path}'.")
set_seed (42)

# === Set device ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

#MODEL PARAMETER COUNTING
def count_params(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

#Code for noise addition
class AddGaussianNoise:
    def __init__(self, mean=0.0, std=0.05):
        self.mean = mean
        self.std = std

    def __call__(self, img):
        if isinstance(img, Image.Image):
            img = np.array(img).astype(np.float32) / 255.0
        noise = np.random.normal(self.mean, self.std, img.shape)
        noisy_img = np.clip(img + noise, 0, 1)
        return Image.fromarray((noisy_img * 255).astype(np.uint8))

class RandomNoiseBlurOrJitter:
    def __init__(self, prob=0.2, noise_std=0.05, blur_radius=1, jitter_params=None):
        self.prob = prob
        self.noise_transform = AddGaussianNoise(std=noise_std)
        self.blur_radius = blur_radius
        self.jitter_transform = ColorJitter(**(jitter_params or {
            'brightness': 0.2,
            'contrast': 0.2,
            'saturation': 0.2,
            'hue': 0.1
        }))

    def __call__(self, img):
        if random.random() < self.prob:
            choice = random.choice(['noise', 'blur', 'jitter'])
            if choice == 'noise':
                img = self.noise_transform(img)
            elif choice == 'blur':
                img = img.filter(ImageFilter.GaussianBlur(radius=self.blur_radius))
            elif choice == 'jitter':
                img = self.jitter_transform(img)
        return img



#Data transformation and loading

# === Data transforms ===
transform_train = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(270),
    RandomNoiseBlurOrJitter(prob=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

transform_val = transforms.Compose([
    transforms.Resize((224, 224)),
    RandomNoiseBlurOrJitter(prob=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])


# === Load datasets ===
train_dataset = datasets.ImageFolder('Primary_Dataset/train', transform=transform_train)
val_dataset = datasets.ImageFolder('Primary_Dataset/val', transform=transform_val)
class_names = train_dataset.classes
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)


# Adding depthwise convolutions after only the first 3 MHSA blocks
import torch
import torch.nn as nn
from timm.models import create_model
from timm.models.vision_transformer import Block, DropPath


# 1. Depthwise convolution module with normalization
class DepthwiseConv(nn.Module):
    def __init__(self, dim, kernel_size=3):
        super().__init__()
        self.dwconv = nn.Conv2d(dim, dim, kernel_size=kernel_size, padding=kernel_size // 2, groups=dim)
        self.bn = nn.BatchNorm2d(dim)

    def forward(self, x):
        B, N, C = x.shape
        cls_token = x[:, :1]
        patch_tokens = x[:, 1:]

        H = W = int(patch_tokens.shape[1] ** 0.5)
        patch_tokens = patch_tokens.transpose(1, 2).view(B, C, H, W)
        patch_tokens = self.bn(self.dwconv(patch_tokens))
        patch_tokens = patch_tokens.flatten(2).transpose(1, 2)

        return torch.cat((cls_token, patch_tokens), dim=1)

# 2. Custom block with parallel MHSA and DepthwiseConv

class ParallelBlock(Block):
    def __init__(self, dim, num_heads, mlp_ratio=4., qkv_bias=True, norm_layer=nn.LayerNorm, drop_path=0.0):
        super().__init__(dim, num_heads, mlp_ratio, qkv_bias=qkv_bias, norm_layer=norm_layer, drop_path=0.0)

        self.depthwise_conv = DepthwiseConv(dim)
        self.attn_norm = norm_layer(dim)
        self.conv_norm = norm_layer(dim)

        # üîÑ Fusion MLP
        self.fuse_mlp = nn.Sequential(
            nn.Linear(dim * 2, dim * 4),
            nn.GELU(),
            nn.Linear(dim * 4, dim)
        )


        self.drop_path = DropPath(drop_path) if drop_path > 0. else nn.Identity()

    def forward(self, x):
        # Parallel MHSA and depthwise conv paths
        x_attn = self.attn(self.attn_norm(x))
        x_conv = self.depthwise_conv(self.conv_norm(x))

        # Fusion
        x_fused = torch.cat([x_attn, x_conv], dim=-1)  # (B, N, 2C)
        x_fused = self.fuse_mlp(x_fused)               # (B, N, C)


        # Residual + MLP
        x = x + self.drop_path(x_fused)
        x = x + self.drop_path(self.mlp(self.norm2(x)))

        return x
    
# 3. Load modified DeiT-tiny with parallel blocks in first 3 layers
def load_custom_parallel_deit(num_classes=200, pretrained=False, replace_blocks=3):
    model = create_model('deit_tiny_patch16_224', pretrained=pretrained)

    for i, block in enumerate(model.blocks):
        if i < replace_blocks:  # Replace first N blocks
            mlp_hidden = block.mlp.fc1.out_features
            mlp_in = block.mlp.fc1.in_features
            mlp_ratio = mlp_hidden / mlp_in

            # Create new ParallelBlock
            new_block = ParallelBlock(
                dim=block.norm1.normalized_shape[0],
                num_heads=block.attn.num_heads,
                mlp_ratio=mlp_ratio,
                qkv_bias=block.attn.qkv.bias is not None,
                norm_layer=type(block.norm1),
                drop_path=0.0
            )

            
            # Replace the block in the model
            model.blocks[i] = new_block
        
    # Replace head for correct number of classes
    model.head = nn.Linear(model.head.in_features, num_classes)
    print(f"‚úÖ Classification head reset for {num_classes} classes.")

    return model
   
# training
#commenting to do gridsearch
# Instantiate and move to device

model = load_custom_parallel_deit(num_classes=200, pretrained=False, replace_blocks=3)
model.load_state_dict(torch.load("TinyImageNet200_pretrained_ConvDeiT-Tiny.pth", map_location=device))

model.head = nn.Linear(model.head.in_features, len(class_names))
model.to(device)
print(f"Total Trainable Parameters in customized  deit tiny model is: {count_params(model):,}")


 # REMOVE TO STOP USING GRID SEARCH
# === Loss and optimizer ===
max_epochs = 80
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=0)
#scheduler=None
scheduler = CosineAnnealingLR (optimizer, T_max=max_epochs)

# === Training with Early Stopping ===
best_acc = 0
epochs = 80
patience = 20
patience_counter = 0
best_epoch = 0

train_losses = []
val_losses = []
train_accuracies = []
val_accuracies = []

for epoch in range(epochs):
    model.train()
    running_loss = 0
    correct_train = 0
    total_train = 0

    for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=False):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

        _, predicted = torch.max(outputs, 1)
        total_train += labels.size(0)
        correct_train += (predicted == labels).sum().item()

    train_loss = running_loss / len(train_loader)
    train_acc = correct_train / total_train

    if scheduler:
        scheduler.step()

    # === Validation ===
    model.eval()
    correct_val = 0
    total_val = 0
    val_running_loss = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)

            val_running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total_val += labels.size(0)
            correct_val += (predicted == labels).sum().item()

    val_loss = val_running_loss / len(val_loader)
    val_acc = correct_val / total_val

    # === Save stats ===
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accuracies.append(train_acc)
    val_accuracies.append(val_acc)

    print(f"Epoch {epoch+1:>3} ‚Äî "
          f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc*100:.2f}% | "
          f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc*100:.2f}%")

    # === Check for improvement ===
    if val_acc > best_acc:
        best_acc = val_acc
        best_epoch = epoch + 1
        patience_counter = 0
        torch.save(model.state_dict(), 
                   'Primary_ValAndTrainingCurvesPloting_ConvDeitTiny.pth')
        print("‚úÖ New best model saved.")
    else:
        patience_counter += 1
        print(f"‚è≥ Patience Counter: {patience_counter}/{patience}")
        if patience_counter >= patience:
            print("üõë Early stopping triggered.")
            break

print(f"\nüéØ Best Accuracy: {best_acc*100:.2f}% at Epoch {best_epoch}")

#Plotting
import matplotlib.pyplot as plt

epochs_range = range(1, len(train_losses) + 1)

# === Plot Accuracy ===
plt.figure(figsize=(4,3))
plt.plot(epochs_range, train_accuracies, label="Training Accuracy")
plt.plot(epochs_range, val_accuracies, label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training vs Validation Accuracy - Primary")
plt.legend()
plt.grid(True)
plt.show()

# === Plot Loss ===
plt.figure(figsize=(4,3))
plt.plot(epochs_range, train_losses, label="Training Loss")
plt.plot(epochs_range, val_losses, label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss - Primary")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import timm
import torch
from torch import nn, optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from tqdm import tqdm
import os
import random
import numpy as np
from PIL import Image, ImageFilter
from torchvision.transforms import ColorJitter
from torch.optim.lr_scheduler import CosineAnnealingLR
from itertools import product
import time
# Customizing DEIT Tiny
import torch
import torch.nn as nn
from timm.models.vision_transformer import VisionTransformer, Block
from timm.models import create_model
from functools import partial
import torch.nn.functional as F
# #for CIFAR-Dataset
# from torchvision.datasets import CIFAR100
# from torch.utils.data import random_split, ConcatDataset
# import torchvision.transforms.functional as F
#Flops and efficiency logging
import time
from ptflops import get_model_complexity_info


# Seed setting
import random
import numpy as np
import torch
import os
import json

# === Reproducibility ===
def set_seed(seed: int = 42, save_path: str = "seed_log.json"):
    """
    Set the random seed for Python, NumPy, and PyTorch.
    Saves the seed to a file so it can be reused later.
    """
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

    # Save seed info
    seed_data = {
        "seed": seed
    }

    with open(save_path, "w") as f:
        json.dump(seed_data, f)

    print(f"[INFO] Seed set to {seed} and logged in '{save_path}'.")
set_seed (42)

# === Set device ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

#MODEL PARAMETER COUNTING
def count_params(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

#Code for noise addition
class AddGaussianNoise:
    def __init__(self, mean=0.0, std=0.05):
        self.mean = mean
        self.std = std

    def __call__(self, img):
        if isinstance(img, Image.Image):
            img = np.array(img).astype(np.float32) / 255.0
        noise = np.random.normal(self.mean, self.std, img.shape)
        noisy_img = np.clip(img + noise, 0, 1)
        return Image.fromarray((noisy_img * 255).astype(np.uint8))

class RandomNoiseBlurOrJitter:
    def __init__(self, prob=0.2, noise_std=0.05, blur_radius=1, jitter_params=None):
        self.prob = prob
        self.noise_transform = AddGaussianNoise(std=noise_std)
        self.blur_radius = blur_radius
        self.jitter_transform = ColorJitter(**(jitter_params or {
            'brightness': 0.2,
            'contrast': 0.2,
            'saturation': 0.2,
            'hue': 0.1
        }))

    def __call__(self, img):
        if random.random() < self.prob:
            choice = random.choice(['noise', 'blur', 'jitter'])
            if choice == 'noise':
                img = self.noise_transform(img)
            elif choice == 'blur':
                img = img.filter(ImageFilter.GaussianBlur(radius=self.blur_radius))
            elif choice == 'jitter':
                img = self.jitter_transform(img)
        return img



#Data transformation and loading

# === Data transforms ===
transform_train = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(270),
    RandomNoiseBlurOrJitter(prob=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

transform_val = transforms.Compose([
    transforms.Resize((224, 224)),
    RandomNoiseBlurOrJitter(prob=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])


# === Load datasets ===
train_dataset = datasets.ImageFolder('CDS_Dataset/train', transform=transform_train)
val_dataset = datasets.ImageFolder('CDS_Dataset/val', transform=transform_val)
class_names = train_dataset.classes
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)


# Adding depthwise convolutions after only the first 2 MHSA blocks
import torch
import torch.nn as nn
from timm.models import create_model
from timm.models.vision_transformer import Block, DropPath


# 1. Depthwise convolution module with normalization
class DepthwiseConv(nn.Module):
    def __init__(self, dim, kernel_size=3):
        super().__init__()
        self.dwconv = nn.Conv2d(dim, dim, kernel_size=kernel_size, padding=kernel_size // 2, groups=dim)
        self.bn = nn.BatchNorm2d(dim)

    def forward(self, x):
        B, N, C = x.shape
        cls_token = x[:, :1]
        patch_tokens = x[:, 1:]

        H = W = int(patch_tokens.shape[1] ** 0.5)
        patch_tokens = patch_tokens.transpose(1, 2).view(B, C, H, W)
        patch_tokens = self.bn(self.dwconv(patch_tokens))
        patch_tokens = patch_tokens.flatten(2).transpose(1, 2)

        return torch.cat((cls_token, patch_tokens), dim=1)

# 2. Custom block with parallel MHSA and DepthwiseConv


class ParallelBlock(Block):
    def __init__(self, dim, num_heads, mlp_ratio=4., qkv_bias=True, norm_layer=nn.LayerNorm, drop_path=0.0):
        super().__init__(dim, num_heads, mlp_ratio, qkv_bias=qkv_bias, norm_layer=norm_layer, drop_path=0.0)

        self.depthwise_conv = DepthwiseConv(dim)
        self.attn_norm = norm_layer(dim)
        self.conv_norm = norm_layer(dim)

        # üîÑ Fusion MLP
        self.fuse_mlp = nn.Sequential(
            nn.Linear(dim * 2, dim * 4),
            nn.GELU(),
            nn.Linear(dim * 4, dim)
        )

        

        self.drop_path = DropPath(drop_path) if drop_path > 0. else nn.Identity()

    def forward(self, x):
        # Parallel MHSA and depthwise conv paths
        x_attn = self.attn(self.attn_norm(x))
        x_conv = self.depthwise_conv(self.conv_norm(x))

        # Fusion
        x_fused = torch.cat([x_attn, x_conv], dim=-1)  # (B, N, 2C)
        x_fused = self.fuse_mlp(x_fused)               # (B, N, C)


        # Residual + MLP
        x = x + self.drop_path(x_fused)
        x = x + self.drop_path(self.mlp(self.norm2(x)))

        return x
    
# 3. Load modified DeiT-tiny with parallel blocks in first 3 layers
def load_custom_parallel_deit(num_classes=len(class_names), pretrained=False, replace_blocks=3):
    model = create_model('deit_tiny_patch16_224', pretrained=pretrained)

    for i, block in enumerate(model.blocks):
        if i < replace_blocks:  # Replace first N blocks
            mlp_hidden = block.mlp.fc1.out_features
            mlp_in = block.mlp.fc1.in_features
            mlp_ratio = mlp_hidden / mlp_in

            # Create new ParallelBlock
            new_block = ParallelBlock(
                dim=block.norm1.normalized_shape[0],
                num_heads=block.attn.num_heads,
                mlp_ratio=mlp_ratio,
                qkv_bias=block.attn.qkv.bias is not None,
                norm_layer=type(block.norm1),
                drop_path=0.0
            )

           # Replace the block in the model
            model.blocks[i] = new_block
        else:
            print(f"‚û°Ô∏è Block {i}: Using original pretrained block.")

    # Replace head for correct number of classes
    model.head = nn.Linear(model.head.in_features, num_classes)
    print(f"‚úÖ Classification head reset for {num_classes} classes.")

    return model
    
# training
#commenting to do gridsearch
# Instantiate and move to device
model = load_custom_parallel_deit(num_classes=len(class_names), pretrained=False, replace_blocks=3)
macs, params = get_model_complexity_info(
    model,
    (3, 224, 224),  # input size
    as_strings=False,   # üëà key: return float values
    print_per_layer_stat=False,
    verbose=False
)

# Convert MACs ‚Üí FLOPs (1 MAC = 2 FLOPs)
flops = macs * 2  

print(f"MACs: {macs:,}")   # commas for readability
print(f"FLOPs: {flops:,}") # expanded number
print(f"Params: {params:,}")

model.to(device)
print(f"Total Trainable Parameters in customized  deit tiny model is: {count_params(model):,}")


# === Loss and optimizer ===
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=0) # changed from lr=1e-5
scheduler = None


# === Training with Early Stopping ===
best_acc = 0
epochs = 200
patience = 20 # changed from 20
patience_counter = 0
best_epoch = 0

for epoch in range(epochs):
    model.train()
    running_loss = 0

    for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=False):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    if scheduler:
        scheduler.step()

    # === Validation ===
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    acc = correct / total
    avg_loss = running_loss / len(train_loader)
    print(f"Epoch {epoch+1:>3} ‚Äî Loss: {avg_loss:.4f}, Val Accuracy: {acc*100:.2f}%")
    

    # === Check for improvement ===
    if acc > best_acc:
        best_acc = acc
        best_epoch = epoch + 1
        patience_counter = 0
        torch.save(model.state_dict(), 'CDS_Scratch_convDeit_tiny.pth')

        print("‚úÖ New best model saved.")
    else:
        patience_counter += 1
        print(f"‚è≥ Patience Counter: {patience_counter}/{patience}")
        if patience_counter >= patience:
            print("üõë Early stopping triggered.")
            break
    
print(f"\nüéØ Best Accuracy: {best_acc*100:.2f}% at Epoch {best_epoch}")
```

```python
import timm
import torch
from torch import nn, optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from tqdm import tqdm
import os
import random
import numpy as np
from PIL import Image, ImageFilter
from torchvision.transforms import ColorJitter
from torch.optim.lr_scheduler import CosineAnnealingLR
from itertools import product
import time
# Customizing DEIT Tiny
import torch
import torch.nn as nn
from timm.models.vision_transformer import VisionTransformer, Block
from timm.models import create_model
from functools import partial
import torch.nn.functional as F
# #for CIFAR-Dataset
# from torchvision.datasets import CIFAR100
# from torch.utils.data import random_split, ConcatDataset
# import torchvision.transforms.functional as F
#Flops and efficiency logging
import time
from ptflops import get_model_complexity_info


# Seed setting
import random
import numpy as np
import torch
import os
import json

# === Reproducibility ===
def set_seed(seed: int = 42, save_path: str = "seed_log.json"):
    """
    Set the random seed for Python, NumPy, and PyTorch.
    Saves the seed to a file so it can be reused later.
    """
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

    # Save seed info
    seed_data = {
        "seed": seed
    }

    with open(save_path, "w") as f:
        json.dump(seed_data, f)

    print(f"[INFO] Seed set to {seed} and logged in '{save_path}'.")
set_seed (42)

# === Set device ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

#MODEL PARAMETER COUNTING
def count_params(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

#Code for noise addition
class AddGaussianNoise:
    def __init__(self, mean=0.0, std=0.05):
        self.mean = mean
        self.std = std

    def __call__(self, img):
        if isinstance(img, Image.Image):
            img = np.array(img).astype(np.float32) / 255.0
        noise = np.random.normal(self.mean, self.std, img.shape)
        noisy_img = np.clip(img + noise, 0, 1)
        return Image.fromarray((noisy_img * 255).astype(np.uint8))

class RandomNoiseBlurOrJitter:
    def __init__(self, prob=0.2, noise_std=0.05, blur_radius=1, jitter_params=None):
        self.prob = prob
        self.noise_transform = AddGaussianNoise(std=noise_std)
        self.blur_radius = blur_radius
        self.jitter_transform = ColorJitter(**(jitter_params or {
            'brightness': 0.2,
            'contrast': 0.2,
            'saturation': 0.2,
            'hue': 0.1
        }))

    def __call__(self, img):
        if random.random() < self.prob:
            choice = random.choice(['noise', 'blur', 'jitter'])
            if choice == 'noise':
                img = self.noise_transform(img)
            elif choice == 'blur':
                img = img.filter(ImageFilter.GaussianBlur(radius=self.blur_radius))
            elif choice == 'jitter':
                img = self.jitter_transform(img)
        return img



#Data transformation and loading

# === Data transforms ===
transform_train = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(270),
    RandomNoiseBlurOrJitter(prob=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

transform_val = transforms.Compose([
    transforms.Resize((224, 224)),
    RandomNoiseBlurOrJitter(prob=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])


# === Load datasets ===
train_dataset = datasets.ImageFolder('CDS_Dataset/train', transform=transform_train)
val_dataset = datasets.ImageFolder('CDS_Dataset/val', transform=transform_val)
class_names = train_dataset.classes
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)


# Adding depthwise convolutions after only the first 2 MHSA blocks
import torch
import torch.nn as nn
from timm.models import create_model
from timm.models.vision_transformer import Block, DropPath


# 1. Depthwise convolution module with normalization
class DepthwiseConv(nn.Module):
    def __init__(self, dim, kernel_size=3):
        super().__init__()
        self.dwconv = nn.Conv2d(dim, dim, kernel_size=kernel_size, padding=kernel_size // 2, groups=dim)
        self.bn = nn.BatchNorm2d(dim)

    def forward(self, x):
        B, N, C = x.shape
        cls_token = x[:, :1]
        patch_tokens = x[:, 1:]

        H = W = int(patch_tokens.shape[1] ** 0.5)
        patch_tokens = patch_tokens.transpose(1, 2).view(B, C, H, W)
        patch_tokens = self.bn(self.dwconv(patch_tokens))
        patch_tokens = patch_tokens.flatten(2).transpose(1, 2)

        return torch.cat((cls_token, patch_tokens), dim=1)

# 2. Custom block with parallel MHSA and DepthwiseConv


class ParallelBlock(Block):
    def __init__(self, dim, num_heads, mlp_ratio=4., qkv_bias=True, norm_layer=nn.LayerNorm, drop_path=0.0):
        super().__init__(dim, num_heads, mlp_ratio, qkv_bias=qkv_bias, norm_layer=norm_layer, drop_path=0.0)

        self.depthwise_conv = DepthwiseConv(dim)
        self.attn_norm = norm_layer(dim)
        self.conv_norm = norm_layer(dim)

        # üîÑ Fusion MLP
        self.fuse_mlp = nn.Sequential(
            nn.Linear(dim * 2, dim * 4),
            nn.GELU(),
            nn.Linear(dim * 4, dim)
        )

        

        self.drop_path = DropPath(drop_path) if drop_path > 0. else nn.Identity()

    def forward(self, x):
        # Parallel MHSA and depthwise conv paths
        x_attn = self.attn(self.attn_norm(x))
        x_conv = self.depthwise_conv(self.conv_norm(x))

        # Fusion
        x_fused = torch.cat([x_attn, x_conv], dim=-1)  # (B, N, 2C)
        x_fused = self.fuse_mlp(x_fused)               # (B, N, C)


        # Residual + MLP
        x = x + self.drop_path(x_fused)
        x = x + self.drop_path(self.mlp(self.norm2(x)))

        return x
    
# 3. Load modified DeiT-tiny with parallel blocks in first 3 layers
def load_custom_parallel_deit(num_classes=len(class_names), pretrained=False, replace_blocks=3):
    model = create_model('deit_tiny_patch16_224', pretrained=pretrained)

    for i, block in enumerate(model.blocks):
        if i < replace_blocks:  # Replace first N blocks
            mlp_hidden = block.mlp.fc1.out_features
            mlp_in = block.mlp.fc1.in_features
            mlp_ratio = mlp_hidden / mlp_in

            # Create new ParallelBlock
            new_block = ParallelBlock(
                dim=block.norm1.normalized_shape[0],
                num_heads=block.attn.num_heads,
                mlp_ratio=mlp_ratio,
                qkv_bias=block.attn.qkv.bias is not None,
                norm_layer=type(block.norm1),
                drop_path=0.0
            )

           # Replace the block in the model
            model.blocks[i] = new_block
        else:
            print(f"‚û°Ô∏è Block {i}: Using original pretrained block.")

    # Replace head for correct number of classes
    model.head = nn.Linear(model.head.in_features, num_classes)
    print(f"‚úÖ Classification head reset for {num_classes} classes.")

    return model
    
# training
#commenting to do gridsearch
# Instantiate and move to device
model = load_custom_parallel_deit(num_classes=len(class_names), pretrained=False, replace_blocks=3)
macs, params = get_model_complexity_info(
    model,
    (3, 224, 224),  # input size
    as_strings=False,   # üëà key: return float values
    print_per_layer_stat=False,
    verbose=False
)

# Convert MACs ‚Üí FLOPs (1 MAC = 2 FLOPs)
flops = macs * 2  

print(f"MACs: {macs:,}")   # commas for readability
print(f"FLOPs: {flops:,}") # expanded number
print(f"Params: {params:,}")

model.to(device)
print(f"Total Trainable Parameters in customized  deit tiny model is: {count_params(model):,}")


# === Loss and optimizer ===
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=0) # changed from lr=1e-5
scheduler = None

import time
import torch
from tqdm import tqdm

# === Training with Early Stopping ===
best_acc = 0.0
epochs = 200
patience = 20
patience_counter = 0
best_epoch = 0

first_epoch_time = None
total_training_time = 0.0
# We'll store each epoch's elapsed time so we can accurately sum up to best_epoch later
epoch_times = []

epochs_completed = 0  # counts how many full epochs we actually ran

for epoch in range(epochs):
    # ---- start epoch timer ----
    if torch.cuda.is_available():
        torch.cuda.synchronize()
    epoch_start = time.perf_counter()
    # ---------------------------

    model.train()
    running_loss = 0.0

    for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=False):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    if scheduler:
        scheduler.step()

    # === Validation ===
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    acc = correct / total
    avg_loss = running_loss / len(train_loader)
    print(f"Epoch {epoch+1:>3} ‚Äî Loss: {avg_loss:.4f}, Val Accuracy: {acc*100:.2f}%")

    # --- Decide improvement ---
    improved = False
    if acc > best_acc:
        best_acc = acc
        best_epoch = epoch + 1
        patience_counter = 0
        improved = True

        # save model
        torch.save(model.state_dict(), 'CDS_Scratch_convDeit_tiny.pth')
        print("‚úÖ New best model saved.")
    else:
        patience_counter += 1
        print(f"‚è≥ Patience Counter: {patience_counter}/{patience}")

    # ---- stop epoch timer ----
    if torch.cuda.is_available():
        torch.cuda.synchronize()
    epoch_end = time.perf_counter()
    epoch_time = epoch_end - epoch_start

    # record times
    epoch_times.append(epoch_time)
    total_training_time += epoch_time
    epochs_completed += 1

    if first_epoch_time is None:
        first_epoch_time = epoch_time

    print(f"‚è±Ô∏è Epoch Time: {epoch_time:.3f} seconds")

    # Early stopping check AFTER timing so final epoch time is included
    if patience_counter >= patience:
        print("üõë Early stopping triggered.")
        break

# End training loop
avg_epoch_time = total_training_time / epochs_completed if epochs_completed > 0 else None

# --- Compute time_to_best_epoch correctly ---
if best_epoch > 0 and len(epoch_times) >= best_epoch:
    # best_epoch is 1-based; sum epoch_times up to and including best_epoch
    time_to_best_epoch = sum(epoch_times[:best_epoch])
else:
    time_to_best_epoch = None

print(f"\nüéØ Best Accuracy: {best_acc*100:.2f}% at Epoch {best_epoch}")

# === Print timing summary ===
print("\n===== TRAINING TIME SUMMARY =====")
print(f"‚è±Ô∏è First Epoch Time:       {first_epoch_time:.3f} seconds" if first_epoch_time is not None else "First epoch time: N/A")
print(f"‚è±Ô∏è Total Training Time:    {total_training_time:.3f} seconds")
print(f"‚è±Ô∏è Avg Epoch Time:         {avg_epoch_time:.3f} seconds" if avg_epoch_time is not None else "Avg epoch time: N/A")

if time_to_best_epoch is not None:
    print(f"‚è±Ô∏è Time to Best Epoch ({best_epoch}): {time_to_best_epoch:.3f} seconds")
else:
    print("‚è±Ô∏è Time to Best Epoch:      No improvement recorded (best never updated)")

print("=================================\n")
```

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import timm
import os
import numpy as np

# === Set device ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# Customizing DEIT Tiny
import torch
import torch.nn as nn
from timm.models.vision_transformer import VisionTransformer, Block
from timm.models import create_model
from functools import partial

# Seed setting
import random
import numpy as np
import torch
import os
import json

# === Reproducibility ===
def set_seed(seed: int = 42, save_path: str = "seed_log.json"):
    """
    Set the random seed for Python, NumPy, and PyTorch.
    Saves the seed to a file so it can be reused later.
    """
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

    # Save seed info
    seed_data = {
        "seed": seed
    }

    with open(save_path, "w") as f:
        json.dump(seed_data, f)

    print(f"[INFO] Seed set to {seed} and logged in '{save_path}'.")
set_seed (42)

# === Set device ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

#MODEL PARAMETER COUNTING
def count_params(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

#Code for noise addition
class AddGaussianNoise:
    def __init__(self, mean=0.0, std=0.05):
        self.mean = mean
        self.std = std

    def __call__(self, img):
        if isinstance(img, Image.Image):
            img = np.array(img).astype(np.float32) / 255.0
        noise = np.random.normal(self.mean, self.std, img.shape)
        noisy_img = np.clip(img + noise, 0, 1)
        return Image.fromarray((noisy_img * 255).astype(np.uint8))

class RandomNoiseBlurOrJitter:
    def __init__(self, prob=0.2, noise_std=0.05, blur_radius=1, jitter_params=None):
        self.prob = prob
        self.noise_transform = AddGaussianNoise(std=noise_std)
        self.blur_radius = blur_radius
        self.jitter_transform = ColorJitter(**(jitter_params or {
            'brightness': 0.2,
            'contrast': 0.2,
            'saturation': 0.2,
            'hue': 0.1
        }))

    def __call__(self, img):
        if random.random() < self.prob:
            choice = random.choice(['noise', 'blur', 'jitter'])
            if choice == 'noise':
                img = self.noise_transform(img)
            elif choice == 'blur':
                img = img.filter(ImageFilter.GaussianBlur(radius=self.blur_radius))
            elif choice == 'jitter':
                img = self.jitter_transform(img)
        return img



#Data transformation and loading

# === Transform for test images (224x224) ===
transform_test = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

# === Load test dataset ===
test_dataset = datasets.ImageFolder('CDS_Dataset/test', transform=transform_test)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Adding depthwise convolutions after only the first 2 MHSA blocks
import torch
import torch.nn as nn
from timm.models import create_model
from timm.models.vision_transformer import Block, DropPath
import torch.nn.functional as F

# 1. Depthwise convolution module with normalization
class DepthwiseConv(nn.Module):
    def __init__(self, dim, kernel_size=3):
        super().__init__()
        self.dwconv = nn.Conv2d(dim, dim, kernel_size=kernel_size, padding=kernel_size // 2, groups=dim)
        self.bn = nn.BatchNorm2d(dim)

    def forward(self, x):
        B, N, C = x.shape
        cls_token = x[:, :1]
        patch_tokens = x[:, 1:]

        H = W = int(patch_tokens.shape[1] ** 0.5)
        patch_tokens = patch_tokens.transpose(1, 2).view(B, C, H, W)
        patch_tokens = self.bn(self.dwconv(patch_tokens))
        patch_tokens = patch_tokens.flatten(2).transpose(1, 2)

        return torch.cat((cls_token, patch_tokens), dim=1)

# 2. Custom block with parallel MHSA and DepthwiseConv



# 2. Custom block with parallel MHSA and DepthwiseConv
class ParallelBlock(Block):
    def __init__(self, dim, num_heads, mlp_ratio=4., qkv_bias=True,
                 norm_layer=nn.LayerNorm, drop_path=0.0):
        super().__init__(dim, num_heads, mlp_ratio,
                         qkv_bias=qkv_bias, norm_layer=norm_layer,
                         drop_path=0.0)

        self.depthwise_conv = DepthwiseConv(dim)
        self.attn_norm = norm_layer(dim)
        self.conv_norm = norm_layer(dim)

        # üëá Learnable gating parameter
        self.gate = nn.Parameter(torch.tensor(0.5))  

        self.drop_path = DropPath(drop_path) if drop_path > 0. else nn.Identity()

    def forward(self, x):
        x_attn = self.attn(self.attn_norm(x))
        x_conv = self.depthwise_conv(self.conv_norm(x))

        # Sigmoid to constrain gate between 0 and 1
        g = torch.sigmoid(self.gate)
        x_fused = g * x_attn + (1 - g) * x_conv

        # Residual
        x = x + self.drop_path(x_fused)

        # MLP + residual
        x = x + self.drop_path(self.mlp(self.norm2(x)))
        return x

# 3. Load modified DeiT-tiny with parallel blocks in first 3 layers
def load_custom_parallel_deit(num_classes=3, pretrained=False, replace_blocks=3):
    model = create_model('deit_tiny_patch16_224', pretrained=pretrained)

    for i, block in enumerate(model.blocks):
        if i < replace_blocks:  # Modify only the first 3 blocks
            mlp_hidden = block.mlp.fc1.out_features
            mlp_in = block.mlp.fc1.in_features
            mlp_ratio = mlp_hidden / mlp_in

            model.blocks[i] = ParallelBlock(
                dim=block.norm1.normalized_shape[0],
                num_heads=block.attn.num_heads,
                mlp_ratio=mlp_ratio,
                qkv_bias=block.attn.qkv.bias is not None,
                norm_layer=type(block.norm1),
                drop_path=0.0
            )

    model.head = nn.Linear(model.head.in_features, num_classes)
    return model
    
    
model = load_custom_parallel_deit(num_classes=3, pretrained=False, replace_blocks=3)
model.load_state_dict(torch.load("CDS_Scratch_GatedSumFusion_3DWCMHSAParallelResidual_ConcatMLP_deit_tiny.pth", map_location=device))
model.to(device)
model.eval()

# === Evaluation ===
all_preds = []
all_labels = []

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        if isinstance(outputs, tuple):
            outputs = outputs[0]
        _, preds = torch.max(outputs, 1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# === Accuracy ===
correct = np.sum(np.array(all_preds) == np.array(all_labels))
total = len(all_labels)
accuracy = 100 * correct / total
print(f"‚úÖ Test Accuracy: {accuracy:.4f}%")

# === Confusion Matrix ===
cm = confusion_matrix(all_labels, all_preds)
class_names = test_dataset.classes

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names,
            yticklabels=class_names)
plt.xlabel('Predicted class')
plt.ylabel('Actual class')
plt.title('Confusion Matrix- deit_tiny_3dwcNorm')
plt.tight_layout()
plt.show()

# === Classification Report ===
report = classification_report(all_labels, all_preds, target_names=class_names, digits=4)
print("üìÑ Classification Report:\n")
print(report)
```

```python
import timm
import torch
from torch import nn, optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from tqdm import tqdm
import os
import random
import numpy as np
from PIL import Image, ImageFilter
from torchvision.transforms import ColorJitter
from torch.optim.lr_scheduler import CosineAnnealingLR
from itertools import product
import time
# Customizing DEIT Tiny
import torch
import torch.nn as nn
from timm.models.vision_transformer import VisionTransformer, Block
from timm.models import create_model
from functools import partial
import torch.nn.functional as F
# #for CIFAR-Dataset
# from torchvision.datasets import CIFAR100
# from torch.utils.data import random_split, ConcatDataset
# import torchvision.transforms.functional as F
#Flops and efficiency logging
import time
from ptflops import get_model_complexity_info


# Seed setting
import random
import numpy as np
import torch
import os
import json

# === Reproducibility ===
def set_seed(seed: int = 42, save_path: str = "seed_log.json"):
    """
    Set the random seed for Python, NumPy, and PyTorch.
    Saves the seed to a file so it can be reused later.
    """
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

    # Save seed info
    seed_data = {
        "seed": seed
    }

    with open(save_path, "w") as f:
        json.dump(seed_data, f)

    print(f"[INFO] Seed set to {seed} and logged in '{save_path}'.")
set_seed (42)

# === Set device ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

#MODEL PARAMETER COUNTING
def count_params(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

#Code for noise addition
class AddGaussianNoise:
    def __init__(self, mean=0.0, std=0.05):
        self.mean = mean
        self.std = std

    def __call__(self, img):
        if isinstance(img, Image.Image):
            img = np.array(img).astype(np.float32) / 255.0
        noise = np.random.normal(self.mean, self.std, img.shape)
        noisy_img = np.clip(img + noise, 0, 1)
        return Image.fromarray((noisy_img * 255).astype(np.uint8))

class RandomNoiseBlurOrJitter:
    def __init__(self, prob=0.2, noise_std=0.05, blur_radius=1, jitter_params=None):
        self.prob = prob
        self.noise_transform = AddGaussianNoise(std=noise_std)
        self.blur_radius = blur_radius
        self.jitter_transform = ColorJitter(**(jitter_params or {
            'brightness': 0.2,
            'contrast': 0.2,
            'saturation': 0.2,
            'hue': 0.1
        }))

    def __call__(self, img):
        if random.random() < self.prob:
            choice = random.choice(['noise', 'blur', 'jitter'])
            if choice == 'noise':
                img = self.noise_transform(img)
            elif choice == 'blur':
                img = img.filter(ImageFilter.GaussianBlur(radius=self.blur_radius))
            elif choice == 'jitter':
                img = self.jitter_transform(img)
        return img



#Data transformation and loading

# === Data transforms ===
transform_train = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(270),
    RandomNoiseBlurOrJitter(prob=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

transform_val = transforms.Compose([
    transforms.Resize((224, 224)),
    RandomNoiseBlurOrJitter(prob=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])


# === Load datasets ===
train_dataset = datasets.ImageFolder('CDS_Dataset/train', transform=transform_train)
val_dataset = datasets.ImageFolder('CDS_Dataset/val', transform=transform_val)
class_names = train_dataset.classes
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)


# Adding depthwise convolutions after only the first 2 MHSA blocks
import torch
import torch.nn as nn
from timm.models import create_model
from timm.models.vision_transformer import Block, DropPath


#Commented code before token merging and prunning
# 1. Depthwise convolution module with normalization
class DepthwiseConv(nn.Module):
    def __init__(self, dim, kernel_size=3):
        super().__init__()
        self.dwconv = nn.Conv2d(dim, dim, kernel_size=kernel_size, padding=kernel_size // 2, groups=dim)
        self.bn = nn.BatchNorm2d(dim)

    def forward(self, x):
        B, N, C = x.shape
        cls_token = x[:, :1]
        patch_tokens = x[:, 1:]

        H = W = int(patch_tokens.shape[1] ** 0.5)
        patch_tokens = patch_tokens.transpose(1, 2).view(B, C, H, W)
        patch_tokens = self.bn(self.dwconv(patch_tokens))
        patch_tokens = patch_tokens.flatten(2).transpose(1, 2)

        return torch.cat((cls_token, patch_tokens), dim=1)

# 2. Custom block with depthwise convolution before the MHSA

class CustomBlock(Block):
    def __init__(self, dim, num_heads, mlp_ratio=4., qkv_bias=True, 
                 norm_layer=nn.LayerNorm, drop_path=0., **kwargs):
        super().__init__(dim, num_heads, mlp_ratio, 
                         qkv_bias=qkv_bias, norm_layer=norm_layer, drop_path=0.0)

        # üëá New: normalization before depthwise conv
        self.norm0 = norm_layer(dim)
        self.depthwise_conv = DepthwiseConv(dim)

        self.drop_path = DropPath(drop_path) if drop_path > 0. else nn.Identity()

    def forward(self, x):
        # Step 1: Depthwise conv with residual (after norm0)
        x = x + self.drop_path(self.depthwise_conv(self.norm0(x)))

        # Step 2: MHSA with residual (after norm1, inherited from Block)
        x = x + self.drop_path(self.attn(self.norm1(x)))

        # Step 3: MLP with residual (after norm2, inherited from Block)
        x = x + self.drop_path(self.mlp(self.norm2(x)))

        return x

    
# 3. Load modified DeiT-tiny with parallel blocks in first 3 layers
def load_custom_parallel_deit(num_classes=3, pretrained=False, replace_blocks=3):
    model = create_model('deit_tiny_patch16_224', pretrained=pretrained)

    for i, block in enumerate(model.blocks):
        if i < replace_blocks:  # Replace first N blocks
            mlp_hidden = block.mlp.fc1.out_features
            mlp_in = block.mlp.fc1.in_features
            mlp_ratio = mlp_hidden / mlp_in

            # Create new ParallelBlock
            new_block = CustomBlock(
                dim=block.norm1.normalized_shape[0],
                num_heads=block.attn.num_heads,
                mlp_ratio=mlp_ratio,
                qkv_bias=block.attn.qkv.bias is not None,
                norm_layer=type(block.norm1),
                drop_path=0.0
            )

           # Replace the block in the model
            model.blocks[i] = new_block
        else:
            print(f"‚û°Ô∏è Block {i}: Using original pretrained block.")

    # Replace head for correct number of classes
    model.head = nn.Linear(model.head.in_features, num_classes)
    print(f"‚úÖ Classification head reset for {num_classes} classes.")

    return model
    
# Instantiate and move to device
model = load_custom_parallel_deit(num_classes=3, pretrained=False, replace_blocks=3)
macs, params = get_model_complexity_info(
    model,
    (3, 224, 224),  # input size
    as_strings=False,   # üëà key: return float values
    print_per_layer_stat=False,
    verbose=False
)

# Convert MACs ‚Üí FLOPs (1 MAC = 2 FLOPs)
flops = macs * 2  

print(f"MACs: {macs:,}")   # commas for readability
print(f"FLOPs: {flops:,}") # expanded number
print(f"Params: {params:,}")

model.to(device)
print(f"Total Trainable Parameters in customized  deit tiny model is: {count_params(model):,}")


# === Loss and optimizer ===
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=0) # changed from lr=1e-5
scheduler = None


# === Training with Early Stopping ===
best_acc = 0
epochs = 200
patience = 20 # changed from 20
patience_counter = 0
best_epoch = 0

for epoch in range(epochs):
    model.train()
    running_loss = 0

    for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=False):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    if scheduler:
        scheduler.step()

    # === Validation ===
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    acc = correct / total
    avg_loss = running_loss / len(train_loader)
    print(f"Epoch {epoch+1:>3} ‚Äî Loss: {avg_loss:.4f}, Val Accuracy: {acc*100:.2f}%")
    

    # === Check for improvement ===
    if acc > best_acc:
        best_acc = acc
        best_epoch = epoch + 1
        patience_counter = 0
        torch.save(model.state_dict(), 'CDS_Scratch_3DWConvBeforeMHSA_deit_tiny.pth')

        print("‚úÖ New best model saved.")
    else:
        patience_counter += 1
        print(f"‚è≥ Patience Counter: {patience_counter}/{patience}")
        if patience_counter >= patience:
            print("üõë Early stopping triggered.")
            break
    
print(f"\nüéØ Best Accuracy: {best_acc*100:.2f}% at Epoch {best_epoch}")
```
